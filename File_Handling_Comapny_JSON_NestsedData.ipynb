```python
import json

with open("/content/sample_data/company_nested_data.json", "r") as file:
  file_lines = json.load(file)

company = file_lines["company"]
name = company["name"]
departments = company["departments"]
locations = company["locations"]

print("Company Name:",name)

for dept in departments:
  dept_id = dept["dept_id"]
  dept_name = dept["dept_name"]
  employees = dept["employees"]

  print("Dept Id:",dept_id)
  print("Department Name:", dept_name)

  for emp in employees:
    emp_id = emp["emp_id"]
    emp_name = emp["name"]
    skills = emp["skills"]
    salary = emp["salary"]

    print("Employee Id:",emp_id)
    print("Employee Name:",emp_name)
    print("Employee Skills:",skills)
    print("Employee Salary:",salary)

print("Company locations:",locations)

```

```python
'''
ðŸ”¹ Question 1: Company Name
Task:
Read the JSON file and print the company name.

Input:
company_data.json

Output:
Company Name
'''
company = file_lines["company"]
name = company["name"]
print("Company Name:", name)
```

```python
'''
ðŸ”¹ Question 2: Total Departments
Task:
Find the total number of departments in the company.

Input:
company_data.json

Output:
Total Departments Count
'''
company = file_lines["company"]
departments = company["departments"]

dept_count = 0

for dept in departments:
  dept_id = dept["dept_id"]

  dept_count += 1

print("Total Departments:", dept_count)
```

```python
'''
ðŸ”¹ Question 3: Department Names List
Task:
Extract and print all department names as a Python list.

Input:
company_data.json

Output:
List of Department Names
'''
company = file_lines["company"]
departments = company["departments"]

python_departments = []

for dept in departments:
  dept_name = dept["dept_name"]
  employees = dept["employees"]

  for emp in employees:
    skills = emp["skills"]

    if "Python" in skills:
      python_departments.append(dept_name)
      break
print(python_departments)
```

```python
'''
ðŸ”¹ Question 4: Total Employees
Task:
Count the total number of employees across all departments.

Input:
company_data.json

Output:
Total Employees Count
'''
company = file_lines["company"]
departments = company["departments"]

total_employees = {}

for dept in departments:
  dept_id = dept["dept_id"]
  employees = dept["employees"]

  for emp in employees:

    if dept_id in total_employees:
      total_employees[dept_id] += 1
    else:
      total_employees[dept_id] = 1

print("The total number of employees across all departments:", total_employees)
```

```python
'''
ðŸ”¹ Question 5: Employees in IT Department
Task:
Print names of all employees working in the IT department.

Input:
company_data.json

Output:
Employee Names (IT Department)
'''
company = file_lines["company"]
departments = company["departments"]

IT_employees = {}

for dept in departments:
  dept_name = dept["dept_name"]
  employees = dept["employees"]

  for emp in employees:
    emp_id = emp["emp_id"]
    emp_name = emp["name"]

    if dept_name == "IT":
      IT_employees[emp_id] = emp_name
print("Employees in IT Department:")

for key, value in IT_employees.items():
  print(f"Employee id: {key}, Employee Name: {value}")
```

```python
'''
ðŸ”¹ Question 6: Employees with Python Skill
Task:
Find and print names of employees who have Python as a skill.

Input:
company_data.json

Output:
Employee Names with Python Skill
'''
company = file_lines["company"]
departments = company["departments"]

python_departments = []

for dept in departments:
  employees = dept["employees"]

  for emp in employees:
    emp_name = emp["name"]
    skills = emp["skills"]

    if "Python" in skills:
      python_departments.append(emp_name)

print(python_departments)
```

```python
'''
ðŸ”¹ Question 7: Average Salary of All Employees
Task:
Calculate the average salary of all employees.

Input:
company_data.json

Output:
Average Salary
'''
company = file_lines["company"]
departments = company["departments"]

emp_count = 0
salary_sum = 0

for dept in departments:
  employees = dept["employees"]

  for emp in employees:
    salary = emp["salary"]

    emp_count += 1
    salary_sum += salary
  avg_salary = salary_sum/emp_count
print(f"The average salary of all employees: {avg_salary:.2f}.")
```

```python
'''
ðŸ”¹ Question 8: Highest Paid Employee
Task:
Find the name and salary of the highest paid employee.

Input:
company_data.json

Output:
Employee Name, Salary
'''
company = file_lines["company"]
departments = company["departments"]

highest_pay = 0
emp_name = ''

for dept in departments:
  employees = dept["employees"]

  for emp in employees:
    name = emp["name"]
    salary = emp["salary"]

    if salary > highest_pay:
      highest_pay = salary
      emp_name = name
print(f"{emp_name} is the highest paid employee with {highest_pay} salary.")
```

```python
'''
ðŸ”¹ Question 9: Department-wise Employee Count
Task:
Create a dictionary showing department name as key and number of employees as value.

Input:
company_data.json

Output:
Dictionary (Department â†’ Employee Count)
'''
employee_department_count = {}

company = file_lines["company"]
departments = company["departments"]

for dept in departments:
  dept_name = dept["dept_name"]
  employees = dept["employees"]

  for emp in employees:
    if dept_name in employee_department_count:
      employee_department_count[dept_name] += 1
    else:
      employee_department_count[dept_name] = 1
print(employee_department_count)
```

```python
'''
ðŸ”¹ Question 10: Company Locations
Task:
Print all company locations in separate lines.

Input:
company_data.json

Output:
Each Location on New Line
'''
company = file_lines["company"]
name = company["name"]
departments = company["departments"]
locations = company["locations"]

print("Company locations:")

for l in locations:
  print(l)
```

```python
'''
ðŸ”¹ Question 11: Employee Name â†’ Department
Task:
Create a dictionary where:

key = employee name (string)

value = department name (string)

Output:
{ 'Amit': 'IT', 'Riya': 'IT', ... }
'''
emp_name_department = {}
company = file_lines["company"]
departments = company["departments"]

for dept in departments:
  dept_name = dept["dept_name"]
  employees = dept["employees"]

  for emp in employees:
    name = emp["name"]

    emp_name_department[name] = dept_name

print(emp_name_department)
```

```python
'''
ðŸ”¹ Question 12: Uppercase Employee Names
Task:
Create a dictionary where:

key = employee ID

value = employee name converted to UPPERCASE

Output:
{1: 'AMIT', 2: 'RIYA', ...}
'''
employee_names = {}
company = file_lines["company"]
departments = company["departments"]

for dept in departments:
  employees = dept["employees"]

  for emp in employees:
    emp_id = emp["emp_id"]
    name = emp["name"]

    employee_names[emp_id] = name
print(employee_names)
```

```python
'''
ðŸ”¹ Question 13: First Character of Employee Names
Task:
Create a dictionary where:

key = employee name

value = first character of the name

Output:
{'Amit': 'A', 'Riya': 'R', ...}
'''
first_char_employee_names = {}
company = file_lines["company"]
departments = company["departments"]

for dept in departments:
  employees = dept["employees"]

  for emp in employees:
    name = emp["name"]

    first_char_employee_names[name] = name[0].upper()
print(first_char_employee_names)
```

```python
'''
ðŸ”¹ Question 14: Name Length Mapping
Task:
Create a dictionary where:

key = employee name

value = length of the employee name

Output:
{'Amit': 4, 'Suresh': 6, ...}
'''
name_length_mapping = {}
company = file_lines["company"]
departments = company["departments"]

for dept in departments:
  employees = dept["employees"]

  for emp in employees:
    name = emp["name"]
    name_length = len(name)

    name_length_mapping[name] = name_length
print(name_length_mapping)
```

```python
'''
ðŸ”¹ Question 15: Department Name â†’ Uppercase
Task:
Create a dictionary where:

key = department name

value = department name in uppercase

Output:
{'IT': 'IT', 'HR': 'HR', 'Finance': 'FINANCE'}
'''
department_name = {}
company = file_lines["company"]
departments = company["departments"]

for dept in departments:
  dept_name = dept["dept_name"]

  department_name[dept_name] = dept_name.upper()
print(department_name)
```

```python
'''
ðŸ”¹ Question 16: Employee Name â†’ Skills String
Task:
Create a dictionary where:

key = employee name

value = all skills joined into a single string separated by commas

Output:
{'Amit': 'Python, SQL', 'Karan': 'Accounting, Excel'}
'''

employee_skills = {}
company = file_lines["company"]
departments = company["departments"]

for dept in departments:
  employees = dept["employees"]

  for emp in employees:
    name = emp["name"]
    skills = emp["skills"]

    if name in employee_skills:
      employee_skills[name].extend(skills)
    else:
      employee_skills[name] = skills

emp_skills_str = {name: ", ".join(skills) for name, skills in employee_skills.items()}
print(emp_skills_str)
```

```python
'''
ðŸ”¹ Question 17: Skill â†’ Employees String
Task:
Create a dictionary where:

key = skill name

value = string of employee names separated by |

Output:
{'Python': 'Amit|Suresh', 'Excel': 'Rahul|Karan'}
'''
skills_emp_name = {}
company = file_lines["company"]
departments = company["departments"]

for dept in departments:
  employees = dept["employees"]

  for emp in employees:
    name = emp["name"]
    skills = emp["skills"]

    for skill in skills:

      if skill in skills_emp_name:
        skills_emp_name[skill].append(name)
      else:
        skills_emp_name[skill] = [name]

emp_name_str = {skills: "| ".join(name) for skills, name in skills_emp_name.items()}
print(emp_name_str)
```

```python
'''
ðŸ”¹ Question 18: Employee Name â†’ Last Character
Task:
Create a dictionary where:

key = employee name

value = last character of the name

Output:
{'Amit': 't', 'Pooja': 'a'}
'''
last_char_employee_names = {}
company = file_lines["company"]
departments = company["departments"]

for dept in departments:
  employees = dept["employees"]

  for emp in employees:
    name = emp["name"]

    last_char_employee_names[name] = name[-1]

print(last_char_employee_names)
```

```python
'''
ðŸ”¹ Question 19: Department Name â†’ Name Length List
Task:
Create a dictionary where:

key = department name

value = list containing lengths of employee names in that department

Output:
{'IT': [4, 4, 6], 'HR': [4, 5]}
'''
department_emp_name = {}
company = file_lines["company"]
departments = company["departments"]

for dept in departments:
  dept_name = dept["dept_name"]
  employees = dept["employees"]

  for emp in employees:
    name = emp["name"]
    name_length = len(name)

    if dept_name in department_emp_name:
      department_emp_name[dept_name].append(name_length)
    else:
      department_emp_name[dept_name] = [name_length]
print(department_emp_name)

```

```python
'''
ðŸ”¹ Question 20: Employee Name â†’ Reversed Name
Task:
Create a dictionary where:

key = employee name

value = reversed string of the employee name

Output:
{'Amit': 'timA', 'Riya': 'ayiR'}
'''
emp_name = {}
company = file_lines["company"]
departments = company["departments"]

for dept in departments:
  employees = dept["employees"]

  for emp in employees:
    name = emp["name"]
    reverse_name = name[::-1]

    emp_name[name] = reverse_name

print(emp_name)
```
