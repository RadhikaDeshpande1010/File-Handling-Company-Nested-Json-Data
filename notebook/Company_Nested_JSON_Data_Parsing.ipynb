Parsing and Formatting Nested JSON Data

```python
import json

# ==========================================================
# ðŸ“Œ Program Name : Read Nested JSON (Company -> Dept -> Emp)
# Purpose        : Extract and display company nested JSON data
# File Used      : company_nested_data.json
# ==========================================================

# Step 1: Load JSON file
with open("/content/sample_data/company_nested_data.json", "r") as json_file:
  json_data = json.load(json_file)

# Step 2: Extract main company details
company_details = json_data["company"]
company_name = company_details["name"]
department_list = company_details["departments"]
company_locations = company_details["locations"]

# Company Basic Details
print("========== COMPANY DETAILS ==========")
print("Company Name:", company_name)

# Step 3: Loop through each department
print("\n========== DEPARTMENT DETAILS ==========")

for department in department_list:

  # Extract department details
  department_id = department["dept_id"]
  department_name = department["dept_name"]
  employee_list = department["employees"]

  print("\n------------------------------")
  print("Dept Id:", department_id)
  print("Department Name:", department_name)

  # Step 4: Loop through employees of the department
  for employee in employee_list:

    # Extract employee details
    employee_id = employee["emp_id"]
    employee_name = employee["name"]
    employee_skills = employee["skills"]
    employee_salary = employee["salary"]

    print("\nEmployee Id:", employee_id)
    print("Employee Name:", employee_name)
    print("Employee Skills:", employee_skills)
    print("Employee Salary:", employee_salary)

# Step 5: Print company locations
print("\n========== COMPANY LOCATIONS ==========")
print("Company locations:", company_locations)
```

```python
"""
==========================================================
ðŸ”¹ Question 1: Company Name
==========================================================

Task:
- Read the JSON file and print the company name.

Input File:
- company_data.json

 Expected Output:
- Company Name
==========================================================
"""
# Extract company details
company_details = json_data["company"]

# Extract company name
company_name = company_details["name"]

# Print company name
print("Company Name:", company_name)
```

```python
"""
==========================================================
ðŸ”¹ Question 2: Total Departments
==========================================================

Task:
- Find the total number of departments in the company.

Input File:
- company_data.json

Expected Output:
- Total Departments Count
==========================================================
"""

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Initialize department count
department_count = 0

# Count total departments
for department in department_list:
  department_count += 1

# Print total department count
print("Total Departments:", department_count)
```

```python
"""
==========================================================
ðŸ”¹ Question 3: Department Names List
==========================================================

Task:
- Extract and print all department names as a Python list.

Input File:
- company_data.json

Expected Output:
- List of Department Names
==========================================================
"""

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# List to store department names where Python skill exists
python_departments_list = []

# Loop through each department
for department in department_list:
  department_name = department["dept_name"]
  employee_list = department["employees"]

  # Loop through each employee inside the department
  for employee in employee_list:
    employee_skills = employee["skills"]

    # Check if "Python" skill is present
    if "Python" in employee_skills:
      python_departments_list.append(department_name)
      break      # Stop checking other employees once Python is found in this department

# Print the final list of departments
print("Departments having at least one employee with Python skill:")
for dept_name in python_departments_list:
  print(dept_name)
```

```python
"""
==========================================================
ðŸ”¹ Question 4: Total Employees
==========================================================

Task:
- Count the total number of employees across all departments.

Input File:
- company_data.json

Expected Output:
- Total Employees Count
==========================================================
"""

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Dictionary to store employee count per department (dept_id : employee_count)
total_employees_count = {}

# Loop through each department
for department in department_list:
  department_id = department["dept_id"]
  employee_list = department["employees"]

  # Loop through each employee in the department
  for employee in employee_list:

    # Count employees under each department_id
    if department_id in total_employees_count:
      total_employees_count[department_id] += 1
    else:
      total_employees_count[department_id] = 1

# Print total employees count
print("The total number of employees across all departments:", total_employees_count)
```

```python
"""
==========================================================
ðŸ”¹ Question 5: Employees in IT Department
==========================================================

Task:
- Print names of all employees working in the IT department.

Input File:
- company_data.json

Expected Output:
- Employee Names (IT Department)
==========================================================
"""

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Dictionary to store IT department employees (emp_id : emp_name)
it_department_employees = {}

# Loop through each department
for department in department_list:
  department_name = department["dept_name"]
  employee_list = department["employees"]

  # Loop through each employee
  for employee in employee_list:
    employee_id = employee["emp_id"]
    employee_name = employee["name"]

    # Store only IT department employees
    if department_name == "IT":
      it_department_employees[employee_id] = employee_name

# Print IT department employee details
print("Employees in IT Department:")

for employee_id, employee_name in it_department_employees.items():
  print(f"Employee id: {employee_id}, Employee Name: {employee_name}")
```

```python
"""
==========================================================
ðŸ”¹ Question 6: Employees with Python Skill
==========================================================

Task:
- Find and print names of employees who have Python as a skill.

Input File:
- company_data.json

Expected Output:
- Employee Names with Python Skill
==========================================================
"""

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# List to store employee names who have Python skill
python_skilled_employees = []

# Loop through each department
for department in department_list:
  employee_list = department["employees"]

  # Loop through each employee
  for employee in employee_list:
    employee_name = employee["name"]
    employee_skills = employee["skills"]

    # Check for Python skill
    if "Python" in employee_skills:
      python_skilled_employees.append(employee_name)

# Print final list of employees
print("Employees with Python Skill:")
for emp_name in python_skilled_employees:
  print(emp_name)
```

```python
"""
==========================================================
ðŸ”¹ Question 7: Average Salary of All Employees
==========================================================

Task:
- Calculate the average salary of all employees.

Input File:
- company_data.json

Expected Output:
- Average Salary
==========================================================
"""

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Initialize employee count and total salary sum
employee_count = 0
total_salary_sum = 0

# Loop through each department
for department in department_list:
  employee_list = department["employees"]

  # Loop through each employee
  for employee in employee_list:
    employee_salary = employee["salary"]

    employee_count += 1
    total_salary_sum += employee_salary

  # Calculate average salary
  average_salary = total_salary_sum / employee_count

# Print average salary of all employees
print(f"The average salary of all employees: {average_salary:.2f}.")
```

```python
"""
==========================================================
ðŸ”¹ Question 8: Highest Paid Employee
==========================================================

Task:
- Find the name and salary of the highest paid employee.

Input File:
- company_data.json

Expected Output:
- Employee Name, Salary
==========================================================
"""

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Initialize highest salary and employee name
highest_salary = 0
highest_paid_employee_name = ""

# Loop through each department
for department in department_list:
  employee_list = department["employees"]

  # Loop through each employee
  for employee in employee_list:
    employee_name = employee["name"]
    employee_salary = employee["salary"]

    # Check for highest salary
    if employee_salary > highest_salary:
      highest_salary = employee_salary
      highest_paid_employee_name = employee_name

# Print final output
print(f"{highest_paid_employee_name} is the highest paid employee with {highest_salary} salary.")
```

```python
"""
==========================================================
ðŸ”¹ Question 9: Department-wise Employee Count
==========================================================

Task:
- Create a dictionary showing department name as key and number of employees as value.

Input File:
- company_data.json

Expected Output:
- Dictionary (Department â†’ Employee Count)
==========================================================
"""

# Dictionary to store department-wise employee count
department_employee_count = {}

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Loop through each department
for department in department_list:
  department_name = department["dept_name"]
  employee_list = department["employees"]

  # Loop through employees in that department
  for employee in employee_list:

    # Increase count for department
    if department_name in department_employee_count:
      department_employee_count[department_name] += 1
    else:
      department_employee_count[department_name] = 1

# Print final dictionary
print("Department-wise Employee Count:")
print(department_employee_count)
```

```python
"""
==========================================================
ðŸ”¹ Question 10: Company Locations
==========================================================

Task:
- Print all company locations in separate lines.

Input File:
- company_data.json

Expected Output:
- Each Location on New Line
==========================================================
"""

# Extract company details
company_details = json_data["company"]

# Extract locations list
company_locations = company_details["locations"]

# Print company locations heading
print("Company locations:")

# Print each location in a new line
for location in company_locations:
  print(location)
```

```python
"""
==========================================================
ðŸ”¹ Question 11: Employee Name â†’ Department
==========================================================

Task:
- Create a dictionary where:
  key   = employee name (string)
  value = department name (string)

Expected Output Example:
{ 'Amit': 'IT', 'Riya': 'IT', ... }
==========================================================
"""

# Dictionary to store employee name and their department name
employee_department_mapping = {}

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Loop through each department
for department in department_list:
  department_name = department["dept_name"]
  employee_list = department["employees"]

  # Loop through each employee in the department
  for employee in employee_list:
    employee_name = employee["name"]

    # Store employee name as key and department name as value
    employee_department_mapping[employee_name] = department_name

# Print employee names and their departments
print("Employee Name â†’ Department Mapping:")
print(employee_department_mapping)
```

```python
"""
==========================================================
ðŸ”¹ Question 12: Uppercase Employee Names
==========================================================

Task:
- Create a dictionary where:
  key   = employee ID
  value = employee name converted to UPPERCASE

Expected Output Example:
{1: 'AMIT', 2: 'RIYA', ...}
==========================================================
"""

# Dictionary to store employee id and employee name (uppercase)
employee_names = {}

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Loop through each department
for department in department_list:
  employee_list = department["employees"]

  # Loop through each employee
  for employee in employee_list:
    employee_id = employee["emp_id"]
    employee_name = employee["name"]

    # Store employee name in uppercase format
    employee_names[employee_id] = employee_name.upper()

# Print employee names in uppercase with employee IDs
print("Employee ID â†’ Uppercase Employee Name Mapping:")
print(employee_names)
```

```python
"""
==========================================================
ðŸ”¹ Question 13: First Character of Employee Names
==========================================================

Task:
- Create a dictionary where:
  key   = employee name
  value = first character of the name

Expected Output Example:
{'Amit': 'A', 'Riya': 'R', ...}
==========================================================
"""

# Dictionary to store employee name and its first character
first_char_employee_names = {}

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Loop through each department
for department in department_list:
  employee_list = department["employees"]

  # Loop through each employee
  for employee in employee_list:
    employee_name = employee["name"]

    # Store first character of employee name in uppercase
    first_char_employee_names[employee_name] = employee_name[0].upper()

# Print employee first-character dictionary
print("Employee Name â†’ First Character Mapping:")
print(first_char_employee_names)
```

```python
"""
==========================================================
ðŸ”¹ Question 14: Name Length Mapping
==========================================================

Task:
- Create a dictionary where:
  key   = employee name
  value = length of the employee name

Expected Output Example:
{'Amit': 4, 'Suresh': 6, ...}
==========================================================
"""

# Dictionary to store employee name and name length
name_length_mapping = {}

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Loop through each department
for department in department_list:
  employee_list = department["employees"]

  # Loop through each employee
  for employee in employee_list:
    employee_name = employee["name"]
    employee_name_length = len(employee_name)

    # Store employee name length
    name_length_mapping[employee_name] = employee_name_length

# Print dictionary of employee names and their lengths
print("Employee Name â†’ Name Length Mapping:")
print(name_length_mapping)
```

```python
"""
==========================================================
ðŸ”¹ Question 15: Department Name â†’ Uppercase
==========================================================

Task:
- Create a dictionary where:
  key   = department name
  value = department name in uppercase

Expected Output Example:
{'IT': 'IT', 'HR': 'HR', 'Finance': 'FINANCE'}
==========================================================
"""

# Dictionary to store department name and uppercase department name
department_uppercase_mapping = {}

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Loop through each department
for department in department_list:
  department_name = department["dept_name"]

  # Store uppercase department name
  department_uppercase_mapping[department_name] = department_name.upper()

# Print dictionary of department names and their uppercase values
print("Department Name â†’ Uppercase Mapping:")
print(department_uppercase_mapping)
```

```python
"""
==========================================================
ðŸ”¹ Question 17: Skill â†’ Employees String
==========================================================

Task:
- Create a dictionary where:
  key   = skill name
  value = string of employee names separated by |

Expected Output Example:
{'Python': 'Amit|Suresh', 'Excel': 'Rahul|Karan'}
==========================================================
"""

# Dictionary to store skill and list of employees having that skill
skill_employee_mapping = {}

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Loop through each department
for department in department_list:
  employee_list = department["employees"]

  # Loop through each employee
  for employee in employee_list:
    employee_name = employee["name"]
    employee_skills = employee["skills"]

    # Loop through each skill
    for skill in employee_skills:

      # Add employee name under the skill
      if skill in skill_employee_mapping:
        skill_employee_mapping[skill].append(employee_name)
      else:
        skill_employee_mapping[skill] = [employee_name]

# Convert list of employee names into string separated by "|"
skill_employee_string_mapping = {
    skill: "| ".join(names) for skill, names in skill_employee_mapping.items()
}

# Print dictionary mapping skills to employee names
print("Skill â†’ Employees String Mapping:")
print(skill_employee_string_mapping)
```

```python
"""
==========================================================
ðŸ”¹ Question 18: Employee Name â†’ Last Character
==========================================================

Task:
- Create a dictionary where:
  key   = employee name
  value = last character of the name

Expected Output Example:
{'Amit': 't', 'Pooja': 'a'}
==========================================================
"""

# Dictionary to store employee name and last character of the name
employee_last_char_mapping = {}

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Loop through each department
for department in department_list:
  employee_list = department["employees"]

  # Loop through each employee
  for employee in employee_list:
    employee_name = employee["name"]

    # Store last character of employee name
    employee_last_char_mapping[employee_name] = employee_name[-1]

# Print last-character dictionary for employee names
print("Employee Name â†’ Last Character Mapping:")
print(employee_last_char_mapping)
```

```python
"""
==========================================================
ðŸ”¹ Question 19: Department Name â†’ Name Length List
==========================================================

Task:
- Create a dictionary where:
  key   = department name
  value = list containing lengths of employee names in that department

Expected Output Example:
{'IT': [4, 4, 6], 'HR': [4, 5]}
==========================================================
"""

# Dictionary to store department name and list of employee name lengths
department_employee_name_length_mapping = {}

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Loop through each department
for department in department_list:
  department_name = department["dept_name"]
  employee_list = department["employees"]

  # Loop through each employee
  for employee in employee_list:
    employee_name = employee["name"]
    employee_name_length = len(employee_name)

    # Store employee name length under department name
    if department_name in department_employee_name_length_mapping:
      department_employee_name_length_mapping[department_name].append(employee_name_length)
    else:
      department_employee_name_length_mapping[department_name] = [employee_name_length]

# Print dictionary of department name and employee name length list
print("Department Name â†’ Employee Name Length List:")
print(department_employee_name_length_mapping)
```

```python
"""
==========================================================
ðŸ”¹ Question 20: Employee Name â†’ Reversed Name
==========================================================

Task:
- Create a dictionary where:
  key   = employee name
  value = reversed string of the employee name

Expected Output Example:
{'Amit': 'timA', 'Riya': 'ayiR'}
==========================================================
"""

# Dictionary to store employee name and reversed employee name
employee_reversed_name_mapping = {}

# Extract company details
company_details = json_data["company"]

# Extract departments list
department_list = company_details["departments"]

# Loop through each department
for department in department_list:
  employee_list = department["employees"]

  # Loop through each employee
  for employee in employee_list:
    employee_name = employee["name"]
    reversed_employee_name = employee_name[::-1]

    # Store employee name and its reversed version
    employee_reversed_name_mapping[employee_name] = reversed_employee_name

# Print dictionary of employee names and their reversed values
print("Employee Name â†’ Reversed Name Mapping:")
print(employee_reversed_name_mapping)
```
